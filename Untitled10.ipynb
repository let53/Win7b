```python

import signal
import sys
import torch
import torch.nn as nn
from torch.utils.data import IterableDataset, DataLoader
from datasets import load_dataset
from collections import Counter
import itertools
import time

# ========== 1. Блокировка остановки ==========
def block_signals():
    signal.signal(signal.SIGINT, signal.SIG_IGN)
    signal.signal(signal.SIGTERM, signal.SIG_IGN)
block_signals()
print("⚠️ Обучение защищено от Ctrl+C и SIGTERM. Остановка только выключением или kill -9", flush=True)

# ========== 2. Настройки ==========
BATCH_SIZE = 4          # уменьшите до 2 при нехватке памяти
BLOCK_SIZE = 256
EMBED_DIM = 256
NUM_HEADS = 8
NUM_LAYERS = 4
DROPOUT = 0.1
LEARNING_RATE = 3e-4
NUM_EPOCHS = 2
TARGET_LOSS = 1
MAX_ARTICLES = None      # None = все статьи русской Википедии (~1.5 млн)
VOCAB_SAMPLE = 50000     # статей для построения словаря (чтобы охватить все символы)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Устройство: {device}", flush=True)

# ========== 3. Построение словаря символов по выборке ==========
print(f"Загрузка выборки из {VOCAB_SAMPLE} статей для построения словаря...", flush=True)
dataset_sample = load_dataset("wikimedia/wikipedia", "20231101.ru", split="train", streaming=True)
char_counter = Counter()
for i, example in enumerate(itertools.islice(dataset_sample, VOCAB_SAMPLE)):
    char_counter.update(example["text"])
    if i % 5000 == 0:
        print(f"Обработано {i} статей для словаря", flush=True)
chars = sorted(char_counter.keys())
vocab_size = len(chars)
stoi = {ch: i for i, ch in enumerate(chars)}
itos = {i: ch for i, ch in enumerate(chars)}
print(f"Размер словаря: {vocab_size} символов", flush=True)

# ========== 4. Предварительный подсчёт общего количества батчей ==========
print("Предварительный подсчёт общего количества батчей (может занять 10-30 минут)...", flush=True)
raw_dataset_count = load_dataset("wikimedia/wikipedia", "20231101.ru", split="train", streaming=True)
total_batches = 0
buffer = []
for i, example in enumerate(raw_dataset_count):
    if MAX_ARTICLES is not None and i >= MAX_ARTICLES:
        break
    text = example["text"]
    indices = [stoi.get(ch, 0) for ch in text if ch in stoi]
    buffer.extend(indices)
    while len(buffer) >= BLOCK_SIZE + 1:
        buffer = buffer[BLOCK_SIZE:]
        total_batches += 1
    if i % 10000 == 0:
        print(f"Обработано {i} статей, батчей накоплено: {total_batches}", flush=True)
print(f"Всего батчей за эпоху: {total_batches:,}", flush=True)

# ========== 5. Потоковый датасет ==========
class StreamingCharDataset(IterableDataset):
    def __init__(self, dataset, stoi, block_size, max_articles):
        self.dataset = dataset
        self.stoi = stoi
        self.block_size = block_size
        self.max_articles = max_articles
    def __iter__(self):
        buffer = []
        for i, example in enumerate(self.dataset):
            if self.max_articles is not None and i >= self.max_articles:
                break
            text = example["text"]
            indices = [self.stoi.get(ch, 0) for ch in text if ch in self.stoi]
            buffer.extend(indices)
            while len(buffer) >= self.block_size + 1:
                x = torch.tensor(buffer[:self.block_size], dtype=torch.long)
                y = torch.tensor(buffer[1:self.block_size+1], dtype=torch.long)
                buffer = buffer[self.block_size:]
                yield x, y

raw_dataset_train = load_dataset("wikimedia/wikipedia", "20231101.ru", split="train", streaming=True)
train_dataset = StreamingCharDataset(raw_dataset_train, stoi, BLOCK_SIZE, MAX_ARTICLES)
train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, num_workers=0)

# ========== 6. Модель ==========
class GPTBlock(nn.Module):
    def __init__(self, embed_dim, num_heads, dropout):
        super().__init__()
        self.ln1 = nn.LayerNorm(embed_dim)
        self.attn = nn.MultiheadAttention(embed_dim, num_heads, dropout=dropout, batch_first=True)
        self.ln2 = nn.LayerNorm(embed_dim)
        self.ff = nn.Sequential(
            nn.Linear(embed_dim, 4 * embed_dim),
            nn.GELU(),
            nn.Linear(4 * embed_dim, embed_dim),
            nn.Dropout(dropout),
        )
        self.dropout = nn.Dropout(dropout)

    def forward(self, x, mask=None):
        x_norm = self.ln1(x)
        attn_out, _ = self.attn(x_norm, x_norm, x_norm, attn_mask=mask, need_weights=False)
        x = x + self.dropout(attn_out)
        x = x + self.dropout(self.ff(self.ln2(x)))
        return x

class MiniGPT(nn.Module):
    def __init__(self, vocab_size, embed_dim, num_heads, num_layers, block_size, dropout):
        super().__init__()
        self.token_emb = nn.Embedding(vocab_size, embed_dim)
        self.pos_emb = nn.Embedding(block_size, embed_dim)
        self.blocks = nn.ModuleList([GPTBlock(embed_dim, num_heads, dropout) for _ in range(num_layers)])
        self.ln_f = nn.LayerNorm(embed_dim)
        self.head = nn.Linear(embed_dim, vocab_size)
        self.block_size = block_size

    def forward(self, x):
        B, T = x.shape
        pos = torch.arange(0, T, device=x.device).unsqueeze(0)
        x = self.token_emb(x) + self.pos_emb(pos)
        mask = torch.triu(torch.ones(T, T, device=x.device) * float('-inf'), diagonal=1)
        for block in self.blocks:
            x = block(x, mask=mask)
        x = self.ln_f(x)
        logits = self.head(x)
        return logits

model = MiniGPT(vocab_size, EMBED_DIM, NUM_HEADS, NUM_LAYERS, BLOCK_SIZE, DROPOUT).to(device)
print(f"Параметров модели: {sum(p.numel() for p in model.parameters()):,}", flush=True)

optimizer = torch.optim.AdamW(model.parameters(), lr=LEARNING_RATE)
criterion = nn.CrossEntropyLoss()

# ========== 7. Обучение с выводом процентов ==========
model.train()
global_step = 0
for epoch in range(NUM_EPOCHS):
    epoch_loss = 0.0
    num_batches_processed = 0
    start_time = time.time()

    last_percent = -1
    for batch_idx, (x, y) in enumerate(train_loader):
        percent = (batch_idx + 1) / total_batches * 100 if total_batches > 0 else 0
        if int(percent) != last_percent:
            last_percent = int(percent)
            print(f"\rEpoch {epoch+1}/{NUM_EPOCHS} — {percent:.1f}%", end="", flush=True)

        x, y = x.to(device), y.to(device)
        logits = model(x)
        loss = criterion(logits.view(-1, vocab_size), y.view(-1))

        optimizer.zero_grad()
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        optimizer.step()

        epoch_loss += loss.item()
        num_batches_processed += 1
        global_step += 1

        if batch_idx % 100 == 0:
            avg_loss_sofar = epoch_loss / num_batches_processed
            print(f"\rEpoch {epoch+1}/{NUM_EPOCHS} — {percent:.1f}% | loss = {loss.item():.4f} (avg: {avg_loss_sofar:.4f})", end="", flush=True)

        if loss.item() < TARGET_LOSS:
            print(f"\nДостигнут целевой loss {TARGET_LOSS} на {percent:.1f}% эпохи. Завершаем обучение досрочно.", flush=True)
            break

    elapsed = time.time() - start_time
    avg_loss = epoch_loss / num_batches_processed if num_batches_processed > 0 else 0
    print(f"\n✅ Эпоха {epoch+1} завершена за {elapsed:.1f} сек. Средний loss: {avg_loss:.4f}", flush=True)

    if loss.item() < TARGET_LOSS:
        break

# ========== 8. Сохранение ==========
print("\nОбучение закончено. Сохраняем модель...", flush=True)
torch.save({
    'model_state_dict': model.state_dict(),
    'vocab_size': vocab_size,
    'embed_dim': EMBED_DIM,
    'num_heads': NUM_HEADS,
    'num_layers': NUM_LAYERS,
    'block_size': BLOCK_SIZE,
    'stoi': stoi,
    'itos': itos,
}, "ru_minigpt_full.pt")
print("Модель сохранена в 'ru_minigpt_full.pt'", flush=True)
sys.exit(0)
```
