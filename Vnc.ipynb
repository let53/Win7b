This notebook sets up a remote desktop environment.

```python
# 1. Install Desktop Environment and VNC
!apt-get update
!apt-get install -y xfce4 xfce4-goodies tightvncserver
```

```python
# 2. Install Bore for tunneling (direct download from GitHub)
!wget https://github.com/ekzhang/bore/releases/download/v0.5.1/bore-v0.5.1-x86_64-unknown-linux-musl.tar.gz
!tar -xf bore-v0.5.1-x86_64-unknown-linux-musl.tar.gz
!cp bore /usr/local/bin/bore
!chmod +x /usr/local/bin/bore
print("Bore installed successfully.")
```

```python
import os

# 3. Setup VNC Password (replace 'password' with your choice)
VNC_PASS = 'colab123'

os.system(f"mkdir -p ~/.vnc")
os.system(f"echo {VNC_PASS} | vncpasswd -f > ~/.vnc/passwd")
os.system(f"chmod 600 ~/.vnc/passwd")

print("VNC Password set to: colab123")
```

```python
# 4. Start VNC and Bore Tunnel
import subprocess
import time
import os

# Set required USER environment variable
os.environ['USER'] = 'root'

# Kill any existing VNC sessions
!vncserver -kill :1 || true
!rm -rf /tmp/.X1-lock /tmp/.X11-unix/X1 || true

# Start VNC server with full path and specific user environment
print("Starting VNC server...")
subprocess.Popen(["vncserver", ":1", "-geometry", "1280x720", "-depth", "24"])

# Wait for VNC to create the socket
time.sleep(5)

# Verify if VNC is listening on 5901
!netstat -tuln | grep 5901

# Start Bore tunnel
print("Starting tunnel... Look for the 'remote_port' in the output below.")
!bore local 5901 --to bore.pub
```

### How to connect:
1. Find the `remote_port` from the bore output above (e.g., `bore.pub:34567`).
2. Open your VNC Viewer client.
3. Address: `bore.pub:XXXXX` (where XXXXX is the port).
4. Password: `colab123`.
